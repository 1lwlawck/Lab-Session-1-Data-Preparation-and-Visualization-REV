**Visualisasi Data**

Scatter Plot

```python
# Scatter plot with salary against number of project

plt.scatter(data_hr['salary'], data_hr['time_spend_company'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Salary')
plt.ylabel('Spend Company')

plt.show()
```

Line Chart

```python
# Plotting line chart with day against tip
plt.plot(data_hr['satisfaction_level'])
plt.plot(data_hr['last_evaluation'])

# Adding Title to the Plot
plt.title("Line Chart")

# Setting the X and Y labels
plt.xlabel('Salary')
plt.ylabel('Number of Project')

plt.show()
```

Bar Chart

```python
# Bar chart with day against tip
plt.bar(data_hr['salary'], data_hr['number_project'])

plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('Salary')
plt.ylabel('Number Project')

# Adding the legends
plt.show()
```

Histogram

```python
# histogram of total_bills
plt.hist(data_hr['last_evaluation'])

plt.title("Histogram")

# Adding the legends
plt.show()
```
